```python
# 1️⃣ Load the dataset using read_csv()
import pandas as pd
df = pd.read_csv('your_dataset.csv')

# 2️⃣ View the first 5 rows of the DataFrame
print("🔹 Head (First 5 rows):")
print(df.head())

# 3️⃣ View the last 3 rows of the DataFrame
print("\n🔹 Tail (Last 3 rows):")
print(df.tail(3))

# 4️⃣ Display 2 random rows from the DataFrame
print("\n🔹 Sample (2 random rows):")
print(df.sample(2))

# 5️⃣ What is the shape, size, and number of dimensions?
print("\n🔹 DataFrame Shape:", df.shape)
print("🔹 DataFrame Size:", df.size)
print("🔹 DataFrame Dimensions:", df.ndim)

# 6️⃣ List all column names and their data types
print("\n🔹 Column Names:", df.columns.tolist())
print("🔹 Data Types:\n", df.dtypes)

# 7️⃣ Use .info() to view structure and explain what it means
print("\n🔹 DataFrame Info:")
df.info()
# Info shows:
# - Total number of rows
# - Column names
# - Non-null values per column
# - Data types
# - Memory usage

# 8️⃣ Use .describe() to see summary statistics for numeric columns
print("\n🔹 Summary Statistics:")
print(df.describe())

# 9️⃣ Set a column (e.g., 'Name') as index
df_indexed = df.set_index('Name')  # Replace 'Name' with actual column if different

# 🔟 Access a row using the new index
print("\n🔹 Access row by index (e.g., 'Sunil'):")
print(df_indexed.loc['Sunil'])  # Replace 'Sunil' with actual index value in your data
```
