## 🧠 What is Pandas?

**Pandas** is an open-source Python library that allows you to work with structured data very efficiently.

It’s especially useful for working with **tabular data**, which is the most common format in ML projects.

With Pandas, you can read, clean, filter, and analyze your datasets in just a few lines of code.


## 🤖 Why Use Pandas in Machine Learning?

Before training any ML model, we must clean and prepare the data.

Pandas helps us:

- Handle missing or incorrect values  
- Rename or modify columns  
- Filter or select data  
- Create new features from existing data

Pandas is your best friend in data preprocessing.

## 📊 Working with Tabular Data

Pandas mainly works with two data structures:

1. **Series** – a one-dimensional labeled array (like a single column)
2. **DataFrame** – a two-dimensional table (like a full Excel sheet)


## 📦 What is a Pandas Series?

A **Series** is a one-dimensional data structure.  
Think of it as a **single column** of data with labels (index).

It stores:
- A list of values
- An index (automatically assigned unless you specify it)


```python
import pandas as pd

# Creating a Series from a list
marks = pd.Series([85, 90, 75])
print(marks)
```

Here,
- `0`, `1`, `2` → are the default **index labels**
- `85`, `90`, `75` → are the actual **data values**


## 🧱 What is a Pandas DataFrame?

A **DataFrame** is a two-dimensional table — like an **Excel sheet**.

It contains:
- **Rows** and **Columns**
- Each column is like a **Series**
- You can have multiple Series combined into a DataFrame

```python
data = {
    'Name': ['John', 'Smith', 'Joe'],
    'Marks': [85, 90, 75]
}

df = pd.DataFrame(data)
print(df)
```
