# 🎥 Video 5: Handling Missing Data in Pandas

In this lesson, we will explore how to handle missing values (NaN/null) in datasets using Pandas.

We'll cover:
- Detecting missing values
- Dropping missing rows/columns
- Filling missing data using mean, median, mode, or custom values
- Real practice on the Titanic dataset


```python
# ✅ Step 1: Import Pandas
import pandas as pd
```

### 📂 Step 2: Load the Titanic Dataset

We are using the Titanic dataset which contains missing values in columns like `Age`, `Cabin`, and `Embarked`.


```python
df = pd.read_csv("Titanic-Dataset.csv")
df.head()
```

### 🔍 Step 3: Check for Missing Values

Start by identifying where and how many missing values exist in your dataset.


```python
# Returns True for missing values
df.isnull()
```

```python
# Total missing values column-wise
df.isnull().sum()
```

```python
# Total missing values in entire dataset
df.isnull().sum().sum()
```

```python
# Percentage of missing values per column
(df.isnull().sum() / len(df)) * 100
```

### 🧹 Step 4: Dropping Missing Data

Remove rows or columns that contain missing data.


```python
# Drop rows with ANY missing values
df.dropna()
```

```python
# Drop columns with ANY missing values
df.dropna(axis=1)
```

```python
# Drop rows where 'Cabin' column is null
df.dropna(subset=['Cabin'])
```

### 🧪 Step 5: Filling Missing Data

Fill missing values with calculated or custom values.


```python
# Fill 'Age' with mean
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
# Fill 'Age' with median
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
# Fill 'Embarked' with mode
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
# Fill 'Cabin' with a custom string
df['Cabin'] = df['Cabin'].fillna("No Info")
```

```python
# Forward fill (fill missing value with previous non-null)
df.fillna(method='ffill')
```

### 🧠 Practice Questions: Handling Missing Data (Titanic Dataset)

1️⃣ Show all rows where `Age` is null.

2️⃣ Drop all rows where `Cabin` is missing.

3️⃣ Fill missing `Age` values with the median.

4️⃣ Fill missing `Embarked` values with the most frequent value (mode).

5️⃣ Fill missing `Fare` values with 0.

6️⃣ Replace missing `Cabin` entries with the string "No Info".

7️⃣ Create a new column `Cabin_missing` that is `True` if `Cabin` is null, else `False`.

8️⃣ Fill all missing values in the DataFrame using forward fill (`ffill`).

9️⃣ Count the total number of missing values in the dataset.

🔟 Drop all columns that contain at least one missing value.


```python
# 1️⃣ Show all rows where Age is null
df[df['Age'].isnull()]

# 2️⃣ Drop all rows where Cabin is missing
df.dropna(subset=['Cabin'])

# 3️⃣ Fill missing Age values with the median
df['Age'] = df['Age'].fillna(df['Age'].median())

# 4️⃣ Fill missing Embarked values with the most frequent value (mode)
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# 5️⃣ Fill missing Fare values with 0
df['Fare'] = df['Fare'].fillna(0)

# 6️⃣ Replace missing Cabin entries with the string "No Info"
df['Cabin'] = df['Cabin'].fillna("No Info")

# 7️⃣ Create a new column Cabin_missing that is True if Cabin is null, else False
df['Cabin_missing'] = df['Cabin'].isnull()

# 8️⃣ Fill all missing values in the DataFrame using forward fill (ffill)
df.fillna(method='ffill')

# 9️⃣ Count the total number of missing values in the dataset
df.isnull().sum().sum()

# 🔟 Drop all columns that contain at least one missing value
df.dropna(axis=1)
```
