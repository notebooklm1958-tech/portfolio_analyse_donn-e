```python
# ✅ Step 1: Import Pandas
import pandas as pd
```

## 📥 Step 2: Create Sample DataFrames
We'll simulate student information and test scores like in real ML datasets.

```python
# Student info
df_students = pd.DataFrame({
    'Student_ID': [101, 102, 103, 104],
    'Name': ['Aarav', 'Bhavya', 'Charan', 'Deepika'],
    'Grade': ['A', 'B', 'A', 'B']
})

# Test scores
df_scores = pd.DataFrame({
    'Student_ID': [101, 102, 104, 105],
    'Math_Score': [90, 85, 88, 95],
    'Science_Score': [92, 89, 84, 90]
})
```

```python
df_students
```

```python
df_scores
```

## 🔗 Step 3: Merging DataFrames using `merge()`

Default is INNER JOIN on common column (`Student_ID`)


```python
# Inner join (only matching rows)
pd.merge(df_students, df_scores, on='Student_ID', how='inner')
```

## 🔄 Step 4: Different Types of Joins


```python
# Left join (keep all students, add matching scores)
pd.merge(df_students, df_scores, on='Student_ID', how='left')
```

```python
# Right join (keep all scores, add matching student info)
pd.merge(df_students, df_scores, on='Student_ID', how='right')
```

```python
# Outer join (keep all rows from both)
pd.merge(df_students, df_scores, on='Student_ID', how='outer')
```

## 📦 Step 5: Concatenating DataFrames using `concat()`

Useful when stacking DataFrames vertically (row-wise) or horizontally (column-wise)


```python
# Vertical stacking (add more student rows)
df_more_students = pd.DataFrame({
    'Student_ID': [106, 107],
    'Name': ['Esha', 'Farhan'],
    'Grade': ['A', 'B']
})

pd.concat([df_students, df_more_students])
```

```python
# Horizontal stacking (add more columns side-by-side)
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'C': [5, 6], 'D': [7, 8]})

pd.concat([df1, df2], axis=1)
```

## 📝 Step 7: Practice Questions

1. Do an inner join between student and score tables.
2. Try outer and left joins – which rows are filled with NaN?
3. Merge on both `Student_ID` and `Grade`.
4. Concatenate two DataFrames row-wise.
5. Stack two DataFrames side-by-side (column-wise).
6. Create a dataset ready for ML by merging info + scores.
7. Add a new column “Total_Score” after merging.


```python
import pandas as pd

# Sample DataFrames
df_students = pd.DataFrame({
    'Student_ID': [101, 102, 103, 104],
    'Name': ['Aarav', 'Bhavya', 'Charan', 'Deepika'],
    'Grade': ['A', 'B', 'A', 'B']
})

df_scores = pd.DataFrame({
    'Student_ID': [101, 102, 104, 105],
    'Score': [88, 77, 91, 83]
})

# 1. Inner Join between student and score tables
inner_join = pd.merge(df_students, df_scores, on='Student_ID', how='inner')
print("1️⃣ Inner Join:\n", inner_join)

# 2. Outer and Left Joins
outer_join = pd.merge(df_students, df_scores, on='Student_ID', how='outer')
left_join = pd.merge(df_students, df_scores, on='Student_ID', how='left')

print("\n2️⃣ Outer Join (NaNs where no match):\n", outer_join)
print("\n2️⃣ Left Join (NaNs for students without scores):\n", left_join)

# 3. Merge on both Student_ID and Grade
# Simulate Grade in df_scores
df_scores_with_grade = pd.DataFrame({
    'Student_ID': [101, 102, 104, 105],
    'Grade': ['A', 'B', 'B', 'C'],
    'Score': [88, 77, 91, 83]
})

merged_on_two = pd.merge(df_students, df_scores_with_grade, on=['Student_ID', 'Grade'], how='inner')
print("\n3️⃣ Merge on Student_ID and Grade:\n", merged_on_two)

# 4. Concatenate two DataFrames row-wise
df_more_students = pd.DataFrame({
    'Student_ID': [106, 107],
    'Name': ['Esha', 'Farhan'],
    'Grade': ['A', 'B']
})
row_concat = pd.concat([df_students, df_more_students], ignore_index=True)
print("\n4️⃣ Row-wise Concatenation:\n", row_concat)

# 5. Stack two DataFrames side-by-side (column-wise)
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'C': [5, 6], 'D': [7, 8]})
col_concat = pd.concat([df1, df2], axis=1)
print("\n5️⃣ Column-wise Concatenation:\n", col_concat)

# 6. Create a dataset ready for ML by merging info + scores
ml_dataset = pd.merge(df_students, df_scores, on='Student_ID', how='inner')

# 7. Add a new column “Total_Score” after merging
ml_dataset['Total_Score'] = ml_dataset['Score']
print("\n6️⃣ & 7️⃣ ML-ready Dataset with Total_Score:\n", ml_dataset)
```
