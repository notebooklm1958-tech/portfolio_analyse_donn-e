# 🏏 IPL Data Analysis – Capstone Project

## 🎯 Project Goal
The goal of this project is to analyze IPL (Indian Premier League) matches dataset to uncover trends, patterns, and insights.  

We will use **NumPy & Pandas** for data analysis and generate meaningful insights about teams, toss decisions, venues, and players.
  
This is a beginner-friendly project without machine learning but gives a complete **real-world sports analytics experience**.


```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("matches.csv")
```

```python
df.head()
```

## 📊 Dataset Overview
Let’s understand the structure of the dataset.


```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['city'] = df['city'].fillna("Unknown")
```

```python
df['player_of_match'] = df['player_of_match'].fillna('No award')
```

```python
df['winner'] = df['winner'].fillna("No result")
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns=['umpire1', 'umpire2'], errors = 'ignore')
```

```python
df.head()
```

## 🔍 Matches per Season


```python
df['season'].value_counts().sort_index()
```

## 🏆 Most Successful Teams


```python
df['winner'].value_counts()
```

## How many matches ended with "No Result"?

```python
(df['winner'] == "No result").sum()
```

## 🎲 Toss Decisions – Does winning toss matter?


```python
toss_match = (df['toss_winner'] == df['winner']).sum()
toss_match
```

```python
total_matches = df.shape[0]
round((toss_match/total_matches)*100)
```

## ⭐ Player of the Match Analysis


```python
df['player_of_match'].value_counts().head(10)
```

## What are the most common toss decisions (bat or field)?

```python
df['toss_decision'].value_counts()
```

## Which teams are most successful at a particular venue? (Example: M Chinnaswamy Stadium)

```python
venue = "Wankhede Stadium"
venue_wins = df[df['venue'] == venue]['winner'].value_counts()
venue_wins
```

## Which matches ended with the smallest and largest margins?

```python
df[df['result'] == 'runs'].sort_values(by='result_margin', ascending=False).head(1)
```

```python
df[df['result'] == 'wickets'].sort_values(by='result_margin', ascending=False).head(1)
```

```python
df[df['result'] == 'runs'].sort_values(by='result_margin', ascending=True).head(1)
```

```python
df[df['result'] == 'wickets'].sort_values(by='result_margin', ascending=True).head(1)
```

## How many matches were decided by Duckworth-Lewis (D/L) method?

```python
df['method'].notnull().sum()
```

## Which team won the most finals?

```python
df[df['match_type'] == 'Final']['winner'].value_counts()
```

# 📌 Key Insights & Conclusion

1. **Matches per Season:** IPL has grown across seasons with varying number of matches.  
2. **Most Successful Teams:** Mumbai Indians and Chennai Super Kings have the highest wins overall.  
3. **Player of the Match:** Legends like AB de Villiers, MS Dhoni, and Chris Gayle have won the most awards.  
4. **No Result Matches:** A few matches ended without results due to rain.  
5. **Toss Impact:** Winning the toss gives a slight advantage but is not a guarantee of winning.  
6. **Venues:** Some stadiums like Wankhede and Eden Gardens host maximum matches and show home team dominance.  
7. **Margins:** IPL has seen nail-biting close wins as well as massive victories.  
8. **D/L Method:** Rarely used but important in rain-affected matches.  
9. **Final Winners:** Mumbai Indians and CSK dominate IPL finals historically.
