In this video, we’ll learn how to **filter data from real datasets** using Pandas — a powerful skill used in every data analysis project.

We'll cover:

🔹 Filter with Logical Conditions: `>`, `<`, `==`, `!=`  
🔹 Combine Conditions with `&` and `|`

Let’s dive into the dataset and filter data smartly using Pandas.

---


```python
# ✅ Step 1: Import Pandas
import pandas as pd
```

📂 **Step 2: Load Real-World Dataset**

We are using a sample student performance dataset.


```python
# Load CSV file
df = pd.read_csv('House_Rent_Dataset.csv')

# Preview the data
df.head()
```

🔍 **Step 3: Filter Using Logical Conditions**


```python
df[df['Rent'] < 100000]
```

🔗 **Step 4: Combining Multiple Conditions with & and |**


```python
df[(df['City']=='Kolkata') | (df['Rent'] > 10000) | (df['Area Locality'] == 'Bandel')]
```

🔤 **Step 5: Filtering Text Columns using .str**


```python
df[df['Area Locality'].str.contains('d')]
```

🔁 **Step 6: Resetting Index After Filtering**

This is helpful to clean up the index after filtering.


```python
f = df[df['City'] == 'Hyderabad']
f.reset_index(inplace=True)
f
```

🚫 **Common Mistakes to Avoid**

✅ Always wrap each condition in parentheses `()`  
✅ Use `&` and `|`, **not** `and`/`or`  
✅ Use `.str` functions only with string columns  
✅ Don’t forget to reset index if needed after filtering


# Practice Questions

These questions help you practice Pandas filtering using conditions, string matching, and logical operators.

---

1️⃣ Show listings with Rent less than ₹10,000

2️⃣ Filter listings that are Semi-Furnished

3️⃣ Show listings with more than 1 Bathroom

4️⃣ Show listings preferred by Bachelors only

5️⃣ Show listings from 'Salt Lake City'

6️⃣ Show Unfurnished listings with Rent below ₹8000

7️⃣ Show listings that are either in 'Dumdum Park' or 'South Dum Dum'

8️⃣ Show listings where Area Type is 'Carpet Area' and Floor starts with '1 out'

9️⃣ Show listings posted after 2022-06-01

🔟 Filter listings where Rent > ₹15,000 and reset the index after filtering


```python
# 1️⃣ Listings with Rent less than ₹10,000
df[df['Rent'] < 10000]

# 2️⃣ Listings that are Semi-Furnished
df[df['Furnishing Status'] == 'Semi-Furnished']

# 3️⃣ Listings with more than 1 Bathroom
df[df['Bathroom'] > 1]

# 4️⃣ Listings preferred by Bachelors only
df[df['Point of Contact'].str.contains('Bachelors', case=False, na=False)]

# 5️⃣ Listings from 'Salt Lake City'
df[df['Area Locality'].str.contains('Salt Lake City', case=False, na=False)]

# 6️⃣ Unfurnished listings with Rent below ₹8000
df[(df['Furnishing Status'] == 'Unfurnished') & (df['Rent'] < 8000)]

# 7️⃣ Listings in 'Dumdum Park' or 'South Dum Dum'
df[df['Area Locality'].isin(['Dumdum Park', 'South Dum Dum'])]

# 8️⃣ Listings with 'Carpet Area' and Floor starts with '1 out'
df[(df['Area Type'] == 'Carpet Area') & (df['Floor'].str.startswith('1 out', na=False))]

# 9️⃣ Listings posted after 2022-06-01
df[pd.to_datetime(df['Posted On']) > pd.to_datetime('2022-06-01')]

# 🔟 Listings with Rent > ₹15,000 and reset index
filtered = df[df['Rent'] > 15000].reset_index(drop=True)
filtered
```
