# 🎥 Video 9: Sorting Data in Pandas

In this video, we'll learn how to **sort data** in Pandas using functions like `sort_values()` and `sort_index()`.

💡 **Why is this important?**

Sorting is very useful in Machine Learning and data analysis. It helps to:
- Organize the data
- Rank or filter top performers
- Prepare datasets in a meaningful order

---

```python
# ✅ Step 1: Import Required Libraries
import pandas as pd
```

## 📥 Step 2: Load Sample Dataset

We'll create a small student dataset.



```python
# Sample student dataset
data = {
    'Student_Name': ['Surya', 'Ravi', 'Anu', 'Divya', 'Kiran'],
    'Math_Score': [85, 92, 78, 90, 88],
    'Science_Score': [89, 95, 80, 85, 91],
    'English_Score': [86, 87, 83, 80, 89]
}

df = pd.DataFrame(data)
df.head()
```

---
## 🔍 Step 3: Sort by One Column

We use `sort_values()` to sort data by a specific column.

**Syntax:**  
`df.sort_values(by='column_name', ascending=True or False)`


```python
# Sort by Math_Score in ascending order
df.sort_values(by='Math_Score', ascending=True)
```

```python
# Sort by Science_Score in descending order
df.sort_values(by='Science_Score', ascending=False)
```

---
## 📌 Step 5: Sorting by Index

Use `sort_index()` when you want to sort by row numbers or labels.


```python
df.sort_index()
```

---
## 🧠 Step 6: Real-Life Use Case – Ranking Students

Let’s calculate each student’s total score and then sort them to rank.


```python
# Add Total_Score column
df['Total_Score'] = df['Math_Score'] + df['Science_Score'] + df['English_Score']

# Sort by Total_Score (high to low)
df_sorted = df.sort_values(by='Total_Score', ascending=False)
df_sorted
```
