```python
# Step 1: Import pandas
import pandas as pd
```

## ✅ Create a Sample DataFrame

1.   List item
2.   List item



We’ll use a small dataset for easy understanding.


```python
data = {
    'Name': ['Joe', 'John', 'Shane'],
    'Subject': ['Maths', 'Science', 'English'],
    'Marks': [85, 91, 78],
    'Passed': [True, True, False]
}

df = pd.DataFrame(data)
df
```

## 📐 Basic Structure

Let’s see the most basic info about rows and columns.


```python
# Shape → (rows, columns)
df.shape
```

```python
# Size → Total number of elements (rows * columns)
df.size
```

```python
# ndim → Number of dimensions (should be 2 for DataFrame)
df.ndim
```

## 🏷️ Labels and Metadata

```python
# Index → Row labels
df.index
```

```python
# Columns → Column labels
df.columns
```

```python
# Axes → Returns both row and column labels
df.axes
```

## 🔢 Data Types & Internal Values

```python
# dtypes → Datatype of each column
df.dtypes
```

```python
# values → Numpy array of all data
df.values
```

## 🧠 Data Inspection Tools

These are used to understand large datasets quickly.

```python
# info() → Structure, non-null count, datatypes
df.info()
```

```python
# describe() → Statistics summary (numeric only)
df.describe()
```

```python
# memory_usage() → Memory used by each column
df.memory_usage()
```

## 🔍 Real Use: Find Datatype Issues

This is useful when cleaning or preparing data.

```python
# Check for object (string) columns
df.select_dtypes(include='object')
```

```python
# Check for numeric columns
df.select_dtypes(include='bool')
```

## 🔓 Accessing Data in a DataFrame

```python
# Access single column
df['Marks']
```

```python
# Access multiple columns
df[['Name', 'Marks']]
```

```python
df
```

```python
# Access row by position using iloc
df.iloc[0:2]
```

```python
# Access row by label using loc
df.loc[0:2]
```

```python
# Access specific cell: 2nd row, 3rd column
df.iloc[1, 2]
```

```python
# Access specific cell by label
df.loc[2, 'Passed']
```

```python
# Access multiple rows and columns
df.loc[0:2, ['Name', 'Subject']]
```

# 🧠 Practice Exercises: DataFrame Properties & Access

Try these questions using the DataFrame you created:

1. What is the shape, size, and number of dimensions of your DataFrame?
2. List all column names and their data types.
3. Use `.info()` and explain what each line means.
4. How much memory does your DataFrame consume?
5. Select only numeric columns using `select_dtypes()`.
6. Access:
   - A single column
   - Two columns
   - The second row
   - A specific cell (row 2, column "Marks")
   - A slice of rows and columns using `.loc`
7. Use `.describe()` to analyze statistics for numeric data.


```python
# 1. What is the shape, size, and number of dimensions?
print("Shape:", df.shape)        # (3, 3) → 3 rows, 3 columns
print("Size:", df.size)          # 9 → Total elements (3x3)
print("Dimensions:", df.ndim)    # 2 → It's a 2D structure (rows & columns)

# 2. List all column names and their data types
print("Columns:", df.columns.tolist())        # ['Name', 'Marks', 'Grade']
print("Data Types:\n", df.dtypes)

# 3. Use .info() and explain each line
df.info()
# Explanation:
# - RangeIndex: Total number of rows.
# - Data columns: Total number of columns and names.
# - Non-null count: Number of non-missing values.
# - Dtype: Data type of each column.
# - Memory usage: Memory consumed by the DataFrame.

# 4. How much memory does it use?
print("Memory usage:")
print(df.memory_usage(deep=True))  # Deep shows object/string memory too

# 5. Select only numeric columns
numeric_df = df.select_dtypes(include='number')
print(numeric_df)

# 6. Access:
print("Single column:\n", df['Marks'])                     # One column
print("Two columns:\n", df[['Name', 'Marks']])             # Two columns
print("Second row:\n", df.iloc[1])                         # Second row (index 1)
print("Cell at row 2, column 'Marks':", df.loc[1, 'Marks'])  # Specific cell

# Access a slice of rows and columns
print("Slice using .loc:\n", df.loc[0:1, ['Name', 'Marks']])

# 7. Use .describe()
print("Summary Statistics:\n", df.describe())
```
