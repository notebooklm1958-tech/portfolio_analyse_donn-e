## ✅ Step 1: Import Required Libraries
We start by importing `pandas` for data analysis.

```python
import pandas as pd
```

## 📥 Step 2: Load the Dataset

We’ll use the messy dataset `StoreData.csv` (which includes missing values, typos, and inconsistencies).
Make sure the file is uploaded to your Colab before running the code below.

```python
df = pd.read_csv("StoreData.csv")
df.head()
```

## 🪪 Step 4: Clean Column Names

Why? Some column names may contain extra spaces, inconsistent casing, or hard-to-read formatting.

Let’s:
- Remove extra spaces
- Convert names to a standard format (Title Case)


```python
df.columns = df.columns.str.strip().str.title().str.replace("_", " ")
df.columns
```

## 🧼 Step 5: Trim Whitespace from Text Data

Why? Sometimes values like `"  Consumer  "` and `"Consumer"` are treated as different.  
We’ll remove extra spaces from strings.


```python
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.strip()
```

## 🔁 Step 6: Remove Duplicate Rows

Why? Duplicate rows can affect analysis like sum, mean, count.

Let’s find and remove them.

```python
print("Before:", len(df))
df = df.drop_duplicates()
print("After:", len(df))
```

## 🔄 Step 7: Convert Data Types

Let’s fix data types for important columns:

- `Order Date` and `Ship Date` → datetime
- `Quantity`, `Sales`, `Profit`, `Discount` → numeric

Also, let’s **remove the % symbol** in the discount column.


```python
# Remove % and convert to float
df['Discount'] = df['Discount'].str.replace('%', '', regex=False)
df['Discount'] = pd.to_numeric(df['Discount'], errors='coerce')

# Convert date columns
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], errors='coerce')

# Convert numeric columns
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')
df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce')
df['Profit'] = pd.to_numeric(df['Profit'], errors='coerce')
```

## 🔎 Step 8: Check for Missing Values After Cleaning

Let’s now see how many values are still missing.

```python
df.isnull().sum()
```

## 🧠 Practice Questions

1. How many rows have missing `Ship Date`?
2. Replace all missing `Profit` values with the median value.
3. Find all rows where `Discount` is greater than 20%.
4. Show the top 5 rows sorted by `Sales` in descending order.
5. Create a new column called `Profit Margin = Profit / Sales`.
6. How many unique `Customer Name` values are there in the dataset?
7. What is the average `Sales` per `Region`?
8. How many orders are from the `Corporate` segment?
9. Find all rows where `Postal Code` contains any letters.
10. Drop all rows where `Order Date`, `Ship Date`, or `Sales` is missing.


```python
# 1. How many rows have missing Ship Date?
print("Missing Ship Date rows:", df['Ship Date'].isna().sum())

# 2. Replace all missing Profit values with the median.
df['Profit'] = df['Profit'].fillna(df['Profit'].median())

# 3. Find all rows where Discount > 20%
high_discount = df[df['Discount'] > 20]
print("Rows with Discount > 20%:")
print(high_discount)

# 4. Show the top 5 rows sorted by Sales in descending order
top_sales = df.sort_values(by='Sales', ascending=False).head(5)
print("Top 5 rows by Sales:")
print(top_sales)

# 5. Create a new column Profit Margin = Profit / Sales
df['Profit Margin'] = df['Profit'] / df['Sales']

# 6. How many unique Customer Name values are there?
unique_customers = df['Customer Name'].nunique()
print("Unique Customer Names:", unique_customers)

# 7. What is the average Sales per Region?
avg_sales_region = df.groupby('Region')['Sales'].mean()
print("Average Sales per Region:")
print(avg_sales_region)

# 8. How many orders are from the Corporate segment?
corporate_orders = df[df['Segment'].str.lower() == 'corporate']
print("Number of Corporate Segment Orders:", len(corporate_orders))

# 9. Find all rows where Postal Code contains any letters
import re
postal_with_letters = df[df['Postal Code'].str.contains(r'[A-Za-z]', na=False)]
print("Postal Codes with letters:")
print(postal_with_letters)

# 10. Drop all rows where Order Date, Ship Date, or Sales is missing
df_cleaned = df.dropna(subset=['Order Date', 'Ship Date', 'Sales'])
print("Remaining rows after dropping missing Order/Ship/Sales:", len(df_cleaned))
```
