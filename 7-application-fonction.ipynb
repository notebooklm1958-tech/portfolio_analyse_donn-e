## 🎯 Step 4: Understanding apply() Function

**What is apply()?** It applies a function to each element in a column or row.

**Basic Syntax:**
- `df['column'].apply(function)` - applies to each element in a column
- `df.apply(function, axis=1)` - applies to each row
- `df.apply(function, axis=0)` - applies to each column

Let's create a simple function to determine pass/fail:

```python
def pass_fail(score):
    if score >= 80:
        return "Pass"
    else:
        return "Fail"

# Apply this function to Math_Score column
df['Math_Result'] = df['Math_Score'].apply(pass_fail)
df[['Student_Name', 'Math_Score', 'Math_Result']].head(10)
```



## 🔥 Step 5: Using Lambda Functions

**What is lambda?** A lambda function is a small, one-line function that doesn't need a name.

**Syntax:** `lambda x: expression`



```python
# Instead of creating a separate function, we can use lambda
df['Science_Result'] = df['Science_Score'].apply(lambda x: "Pass" if x >= 80 else "Fail")
df[['Student_Name', 'Science_Score', 'Science_Result']].head(10)
```

### Convert Scores to Letter Grades

```python
# Convert scores to letter grades using lambda
df['Math_Grade'] = df['Math_Score'].apply(
    lambda x: 'A' if x >= 90 else 'B' if x >= 80 else 'C' if x >= 70 else 'D' if x >= 60 else 'F'
)

# Check if student is adult (age >= 21)
df['Is_Adult'] = df['Age'].apply(lambda x: "Yes" if x >= 21 else "No")

df[['Student_Name', 'Math_Score', 'Math_Grade', 'Age', 'Is_Adult']].head(10)
```

## 🗺️ Step 6: Using map() Function

**What is map()?** Similar to apply() but works great with dictionaries for value mapping.

```python
# Create a mapping dictionary
grade_mapping = {
    'A': 'Excellent',
    'B': 'Good',
    'C': 'Average',
    'D': 'Below Average',
    'F': 'Poor'
}

# Use map() to convert grades to descriptions
df['Grade_Description'] = df['Math_Grade'].map(grade_mapping)
df[['Student_Name', 'Math_Grade', 'Grade_Description']].head(10)
```

## 📊 Step 7: Apply Functions to Multiple Columns (Row-wise)

Sometimes we need to use multiple columns together to create new data.

```python
# Calculate average score for each student
def calculate_average(row):
    return (row['Math_Score'] + row['Science_Score'] + row['English_Score']) / 3

df['Average_Score'] = df.apply(calculate_average, axis=1)

# Calculate total score using lambda
df['Total_Score'] = df.apply(lambda row: row['Math_Score'] + row['Science_Score'] + row['English_Score'], axis=1)

df[['Student_Name', 'Math_Score', 'Science_Score', 'English_Score', 'Average_Score', 'Total_Score']].head(10)
```

## 🧠 Practice Questions

1. How many students have **Outstanding** performance?
2. Create a column that shows **"High Scorer"** for students with total score > 240
3. Find students who are **eligible** for scholarship
4. Calculate the **score range** (difference between highest and lowest score) for each student
5. Create **age categories**: Young (< 21), Adult (21-23), Mature (> 23)
6. Count how many students have **Math** as their best subject
7. Create a **consistency check**: "Consistent" if score difference ≤ 10, else "Inconsistent"
8. Find students from **Senior** grade level with **average score >= 88**

```python
import pandas as pd

# Sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Grade': ['Senior', 'Junior', 'Senior', 'Sophomore', 'Senior'],
    'Age': [22, 20, 24, 21, 23],
    'Math': [85, 92, 88, 78, 91],
    'Science': [90, 89, 84, 80, 87],
    'English': [88, 85, 91, 75, 92],
    'Total': [263, 266, 263, 233, 270],
    'Performance': ['Outstanding', 'Good', 'Outstanding', 'Average', 'Outstanding']
}

df = pd.DataFrame(data)

# Q1. How many students have Outstanding performance?
outstanding_count = df[df['Performance'] == 'Outstanding'].shape[0]
print("Q1 - Outstanding Students:", outstanding_count)

# Q2. Create a column "High Scorer" for total score > 240
df['High Scorer'] = df['Total'] > 240
print("\nQ2 - High Scorer Column:\n", df[['Name', 'Total', 'High Scorer']])

# Q3. Find students who are eligible for scholarship (Assume eligible if Total > 250)
eligible_scholarship = df[df['Total'] > 250]
print("\nQ3 - Eligible for Scholarship:\n", eligible_scholarship[['Name', 'Total']])

# Q4. Score range (max - min score of 3 subjects)
df['Score Range'] = df[['Math', 'Science', 'English']].max(axis=1) - df[['Math', 'Science', 'English']].min(axis=1)
print("\nQ4 - Score Range:\n", df[['Name', 'Score Range']])

# Q5. Age Categories
df['Age Category'] = pd.cut(df['Age'], bins=[0, 20, 23, float('inf')], labels=['Young', 'Adult', 'Mature'])
print("\nQ5 - Age Categories:\n", df[['Name', 'Age', 'Age Category']])

# Q6. Count students who have Math as their best subject
df['Best Subject'] = df[['Math', 'Science', 'English']].idxmax(axis=1)
math_best_count = df[df['Best Subject'] == 'Math'].shape[0]
print("\nQ6 - Math as Best Subject Count:", math_best_count)

# Q7. Consistency check (score difference ≤ 10 between highest and lowest)
df['Consistency'] = df['Score Range'].apply(lambda x: 'Consistent' if x <= 10 else 'Inconsistent')
print("\nQ7 - Consistency:\n", df[['Name', 'Score Range', 'Consistency']])

# Q8. Students from Senior grade level with average score >= 88
df['Average Score'] = df[['Math', 'Science', 'English']].mean(axis=1)
senior_high_avg = df[(df['Grade'] == 'Senior') & (df['Average Score'] >= 88)]
print("\nQ8 - Senior with Average >= 88:\n", senior_high_avg[['Name', 'Grade', 'Average Score']])
```
